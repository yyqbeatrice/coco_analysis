```python
## summarize
import openai
import os


round = "round0"
meeting_id = "10"
user = "user6"
speaker = 'speaker2'
prefer_path = f'summary/group2/{round}/prefer/{user}_{meeting_id}_prefer.txt'
file_id = f'{round}_{meeting_id}_{speaker}'

fixed_file = f'summary/group2/{round}/{round}_{meeting_id}/{file_id}_fixed.txt'
fixed_output = open(fixed_file, "r").read()

raw_transcript_file_path = f'transcript/group2new/{user}/{meeting_id}.txt'
raw_transcript = open(raw_transcript_file_path, "r").read()
original_output = open(f"summary/group2/{round}/{round}_{meeting_id}/{file_id}.txt", "r").read()



if os.path.exists(prefer_path):
    preference = open(prefer_path, "r").read()
else:
    preference = ""

instruction = f"""
根据以下一次会议的内容以及prompt要求，AI为{speaker}定制了一份笔记
{speaker}根据自己的偏好对笔记进行了修正
比较修改前后的笔记，分析用户对笔记的偏好
结合该用户历史偏好，重新总结其偏好
如果历史偏好为空，则直接根据笔记分析即可

会议内容:{raw_transcript}
AI笔记: {original_output}
用户修正后的笔记：{fixed_output}
用户历史偏好：{preference}
"""

completion = openai.chat.completions.create(
    model="gpt-4",
    messages=[
        {"role": "system", "content": "你是一个文本分析专家。"},
        {"role": "user", "content": instruction}
    ],
    temperature=0
)
out =  completion.choices[0].message.content

print(out)
print(prefer_path)
with open(prefer_path, "w") as text_file:
    text_file.write(out)
```

```python
## summarize
import openai
import os

# text_type = "transcript"
round = "round1"
meeting_id = "12"
user = "user5"
speaker = 'speaker2'
prefer_path = f'summary/group2/{round}/prefer/{user}_{meeting_id}_prefer.txt'
preference = open(prefer_path, "r").read()

cur_meeting_id = '12'
cur_speaker = 'speaker2'
cur_raw_transcript_file_path = f'transcript/group2new/{user}/{cur_meeting_id}.txt'
cur_raw_transcript = open(raw_transcript_file_path, "r").read()

cur_round = "round1"
cur_file_id = f'{cur_round}_{cur_meeting_id}_{cur_speaker}'
output_answer_file_path = f'summary/group2/{cur_round}/{cur_round}_{cur_meeting_id}/{cur_file_id}.txt'

if not os.path.exists(output_answer_file_path):
    os.makedirs(os.path.dirname(output_answer_file_path), exist_ok=True)


instruction = f"""
请根据{cur_speaker}偏好为其定制一份中文笔记

要求：
1.生成标题概括主题，标题前加上'# '号
2.概述会议的主要内容和待办事项

{cur_speaker}偏好：{preference}

会议记录：
{raw_transcript}
"""

print(instruction)

completion = openai.chat.completions.create(
    model="gpt-4",
    messages=[
        {"role": "system", "content": "你是一名笔记助手，擅长总结会议记录。"},
        {"role": "user", "content": instruction}
    ],
    temperature=0
)
out =  completion.choices[0].message.content
print(out)
with open(output_answer_file_path, "w") as text_file:
    text_file.write(out)
```
